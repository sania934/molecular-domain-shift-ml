```python
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
```

```python
import pandas as pd

qm9 = pd.read_csv("qm9_5k.csv")
hopv = pd.read_csv("hopv_clean_target_prop7.csv")

print(qm9.shape)
print(hopv.shape)
```

```python
#Morgan fingerprint function
def mol_fp(smiles, n_bits=1024, radius=2):
    mol = Chem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
    
    arr = np.zeros((n_bits,))
    AllChem.DataStructs.ConvertToNumpyArray(fp, arr)
    return arr
```

```python
#Apply to datasets
qm9["fp"] = qm9["smiles"].apply(mol_fp)
hopv["fp"] = hopv["SMILES"].apply(mol_fp)
```

```python
#Save processed datasets
qm9.to_csv("qm9_5k_processed.csv", index=False)
hopv.to_csv("hopv_processed.csv", index=False)
```

```python
#Convert to arrays
import numpy as np

X_qm9 = np.stack(qm9["fp"])
y_qm9 = qm9["gap"]

X_hopv = np.stack(hopv["fp"])
y_hopv = hopv["prop_7"]
```

```python
np.save("X_qm9.npy", X_qm9)
np.save("y_qm9.npy", y_qm9)

np.save("X_hopv.npy", X_hopv)
np.save("y_hopv.npy", y_hopv)
```
