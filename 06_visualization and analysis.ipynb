```python
#load parameters
import numpy as np

X_qm9 = np.load("X_qm9.npy")
y_qm9 = np.load("y_qm9.npy")

X_hopv = np.load("X_hopv.npy")
y_hopv = np.load("y_hopv.npy")
```

```python
#PCA 
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X_all = np.concatenate([X_qm9, X_hopv])
labels = ["QM9"] * len(X_qm9) + ["HOPV"] * len(X_hopv)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_all)

plt.scatter(X_pca[:,0], X_pca[:,1], c=[0]*len(X_qm9)+[1]*len(X_hopv))
plt.title("QM9 vs HOPV Chemical Space")
plt.show()
```
